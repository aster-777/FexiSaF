```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
df.columns
```

```python
df_total_profit = df['total_profit']
df_month = df['month_number']
```

```python
plt.plot(df_month,df_total_profit, marker='o', linestyle='-', color= 'orange')
plt.xlabel('Month')
plt.ylabel('Total profit')
plt.title('Monthly Total Profit')
plt.show()
```
